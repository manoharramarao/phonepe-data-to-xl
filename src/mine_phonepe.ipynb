```python
# 1. Aggregated Transactions by type
converted_trans_data_state = []
for trans_file_state in glob.iglob('./data/aggregated/transaction/country/india' + '/**/*.json', recursive=True):
    if(not re.search('state', trans_file_state)):
        file_split = trans_file_state.split('/')
        year = file_split[6]
        quarter = file_split[7].split('.')[0]
        with open(trans_file_state) as json_file:
            data = json.load(json_file)
        data = json.loads(json.dumps(data), object_hook=lambda d: SimpleNamespace(**d))
        for trans_data_state in data.data.transactionData:
            temp_trans_data_state = {}
            temp_trans_data_state['country'] = 'IN'
            temp_trans_data_state['year'] = year
            temp_trans_data_state['quarter'] = quarter
            temp_trans_data_state['type'] = trans_data_state.name
            temp_trans_data_state['amount'] = trans_data_state.paymentInstruments[0].amount
            temp_trans_data_state['count'] = trans_data_state.paymentInstruments[0].count
            converted_trans_data_state.append(temp_trans_data_state)
df = pd.DataFrame(converted_trans_data_state)
df.to_excel('./data/converted/01_agg_trans_by_type.xlsx')
```
